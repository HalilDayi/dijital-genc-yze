```python
import pandas
```

```python
data = pandas.read_csv('data/Crop_Recommendation.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
train_data = data.drop(['Crop'], axis=1)
test_data = data['Crop']
```

```python
train_data.describe()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

sc = StandardScaler()
train_data_stan = sc.fit_transform(train_data)
```

```python
train_data_stan_df = pandas.DataFrame(data=train_data_stan, columns=train_data.columns)

train_data_stan_df.describe()
```

```python
from sklearn.cluster import KMeans

km_model = KMeans(n_clusters=22)
km_model.fit(train_data_stan)
clusters = km_model.predict(train_data_stan)
```

```python
clusters
```

```python
max(clusters)
```

```python
min(clusters)
```

```python
len(clusters)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

le.fit(test_data)

test_data_enc = le.transform(test_data)

test_data_enc
```

```python
distribution = []

for x in range(max(clusters) + 1):
    distribution.append([])

for x in range(len(clusters)):
    distribution[clusters[x]].append(test_data_enc[x])
```

```python
distribution
```

```python
import numpy

arr = numpy.zeros(shape=[22, 22])
```

```python
arr
```

```python
for index, cluster in enumerate(distribution):
    for crop in cluster:
        arr[index, crop] += 1
```
